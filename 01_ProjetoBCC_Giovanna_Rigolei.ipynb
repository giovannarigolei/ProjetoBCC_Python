#Variáveis

Para esta análise usaremos as seguintes variáveis abaixo.

**Variáveis:**

**1 - Taxa de desemprego no Brasil.**

A taxa de desemprego se refere às pessoas com idade para trabalhar que não estão trabalhando, mas estão disponíveis e tentam encontrar trabalho.

**2 - Variação do PIB do Brasil.**

O PIB (Produto Interno Bruto) pode ser definido como a soma de todos os bens e serviços finais produzidos por um país, estado ou cidade.

**3 - Renda média mensal.**

A Renda média mensal é o valor médio recebido por todas as pessoas que têm algum tipo de rendimento no Brasil.


**Frequência de ambas as variáveis:** Trimestral.

**Período analisado de ambas as variáveis:** Do 4° Trimestre de 2017 ao 1° Trimestre de 2020.

**Fonte de ambas as variáveis:** https://www.ibge.gov.br/indicadores.html

#Relacionando os Dados

**Relação entre a taxa de desemprego e as variações do PIB:**

Correlação e Regressão entre a Taxa de Desemprego e as Variações do PIB:


```
#Correlação
corr=taxdes['Taxa de Desemprego'].corr(pb['Variação do PIB'])
if corr<0:
  corrF=corr*(-1)
  if corrF == 0 or corrF <= 0.19:
    intensidade=". Correlação negativa muito fraca."
  elif corrF > 0.19 and corrF <= 0.39:
    intensidade=". Correlação negativa fraca."
  elif corrF > 0.39 and corrF <= 0.69:
    intensidade=". Correlação negativa moderada."
  elif corrF > 0.69 and corrF <= 0.89:
    intensidade=". Correlação negativa forte."
  elif corrF > 0.89 and corrF <= 1:
    intensidade=". Correlação negativa muito forte."
if corr>=0:
  corrF=corr
  if corrF == 0 or corrF <= 0.19:
    intensidade=". Correlação positiva muito fraca."
  elif corrF > 0.19 and corrF <= 0.39:
    intensidade=". Correlação positiva fraca."
  elif corrF > 0.39 and corrF <= 0.69:
    intensidade=". Correlação positiva moderada."
  elif corrF > 0.69 and corrF <= 0.89:
    intensidade=". Correlação positiva forte."
  elif corrF > 0.89 and corrF <= 1:
    intensidade=". Correlação positiva muito forte."

print("Coeficiente de Correlação entre as variáveis:", corr, intensidade)
print("Reta de Regressão: ")

#Regressão
m, b = np.polyfit(taxdes['Taxa de Desemprego'], pb['Variação do PIB'], deg = 1)
plt.scatter(taxdes['Taxa de Desemprego'], pb['Variação do PIB'], 30, color='red')
ax = plt.gca()
ax.set_xlabel("Taxa de Desemprego (x)")
ax.set_ylabel("Variação do PIB (y)")
x = taxdes['Taxa de Desemprego']
y = m*x + b
plt.plot(x,y, linewidth=2.0, linestyle="-", color="black")
plt.grid()
plt.show()
```

Qual foi a diferença entre a variação média porcentual (total) da **Taxa de Desemprego** e a do **PIB** durante todo o período analisado?



```
taxa=taxdes["Taxa de Desemprego"]
a=0
b=1
difs=[]
while b<10:
  d=[(taxa[b] - taxa[a]) * 100/ (taxa[a])]
  difs=difs+d
  a=b
  b=b+1
x=0
a=0
c=0
while x < 9 :
  if difs[x]>0:
    a = a + difs[x]
    x=x+1
  else:
    b=difs[x]*(-1)
    c = c + b
    x=x+1
mediaT=(a+c)/9


pib=pb["Variação do PIB"]
a=0
b=1
difs=[]
while b<10:
  d=[(pib[b] - pib[a]) * 100/ (pib[a])]
  difs=difs+d
  a=b
  b=b+1

x=0
a=0
c=0
while x < 9 :
  if difs[x]>0:
    a = a + difs[x]
    x=x+1
  else:
    b=difs[x]*(-1)
    c = c + b
    x=x+1
mediaP=(a+c)/9

DI=mediaP-mediaT
print("A variação porcentual média da Taxa de desemprego foi de:", "%.2f" % mediaT, "%.")
print("A variação porcentual média do PIB foi de:", "%.2f" % mediaP, "%.")
print("A diferença entre as variações médias porcentuais entre a Taxa de Desemprego e o PIB durante todo o período analisado é de:", "%.2f" % DI, "%.")
```

Em quais trimestres a relação entre a **Taxa de Desemprego** e a **Renda Média Mensal** foi positiva no ponto de vista econômico e social? Ou seja, em quais trimestres dentre os analisados, a taxa de desemprego diminuiu ao mesmo tempo em que a renda média mensal aumentou?

Como sabemos por meio dos gráficos que a Renda Média Mensal Brasileira sofreu apenas aumentos durante o período de tempo analisado, precisamos apenas descobrir os trimestres em que a a Taxa de Desemprego diminuiu.


```
tx=taxdes["Taxa de Desemprego"]
td=taxdes["Ano e Trimestre"]

a=0
b=1
x=0
trimestre=[]

while x<9:
  if tx[b] < tx[a]:
      t=[x+1]
      trimestre=trimestre+t
      x=x+1
      a=b
      b=b+1
  else:
      x=x+1
      a=b
      b=b+1


numeroDeTrimestre=len(trimestre)
n=numeroDeTrimestre-1
a=0
print("A relação entre a Taxa de Desemprego e a Renda Média Mensal foi positiva no ponto de vista econômico e social nas seguintes datas:")
while a<=n:
  print(td[a])
  a=a+1

```

As variações do **PIB** foram proporcionais às variações da **Renda Média Mensal** brasileira durante o período analisado?

```
pib=pb["Variação do PIB"]
rm=ren["Renda"]

a=0
b=1
x=0
p=0

pib[b]*rm[b]

while x < 9:
  if pib[a]/rm[a] == pib[b]/rm[b]:
    p=p+1
    a=b
    b=b+1
    x=x+1
  elif pib[a]*rm[a] == pib[b]*rm[b]:
    p=p-1
    a=b
    b=b+1
    x=x+1
  else:
    x=x+1


if p!=9 or p!=-9:
  prop="As variações do PIB e da Renda Média Mensal brasileira durante o período analisado, não foram proporcionais."
elif p==9:
  prop="As variações do PIB e da Renda Média Mensal brasileira durante o período analisado, foram diretamente proporcionais."
elif p==-9:
  prop="As variações do PIB e da Renda Média Mensal brasileira durante o período analisado, foram inversamente proporcionais."

print(prop)
```
